```python
import pandas as pd

df = pd.read_csv("startup_funding.csv")
df.head()
```

```python
df.columns
```

```python
# Triming the Colums Names
df.columns = df.columns.str.strip()          # remove extra spaces at start/end
df.columns = df.columns.str.replace('  ', ' ')  # replace double spaces
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
#Check Missing Values
df.isnull().sum()
```

```python
# Handle missing values
df['industry_vertical'] = df['industry_vertical'].fillna("Unknown")
df['city_location'] = df['city_location'].fillna("Unknown")
df['investors_name'] = df['investors_name'].fillna("Undisclosed")

# Convert amount column
df['amount_in_usd'] = pd.to_numeric(df['amount_in_usd'], errors='coerce')
df['amount_in_usd'] = df['amount_in_usd'].fillna(df['amount_in_usd'].median())
```

```python
#Remove Duplicate
df.drop_duplicates(inplace=True)
```

```python
#Clean Text Columns
df['startup_name'] = df['startup_name'].str.strip().str.title()
df['city_location'] = df['city_location'].str.strip().str.title()
df['industry_vertical'] = df['industry_vertical'].str.strip().str.title()
df['investors_name'] = df['investors_name'].str.strip()
```

```python
#Fix Date & Year
df['date_dd/mm/yyyy'] = pd.to_datetime(df['date_dd/mm/yyyy'], errors='coerce')
df['year'] = df['date_dd/mm/yyyy'].dt.year
```

```python
#Create New Column
df['funding_category'] = df['amount_in_usd'].apply(lambda x:
    'Low' if x < 1000000 else
    'Medium' if x < 10000000 else
    'High'
)
```

```python
#cleaned_dataset
df.to_csv("cleaned_dataset.csv", index=False)
```
